En este segundo notebook lo que queremos hacer es profundizar en las columnas que hemos elegido, eliminando todos los los valores nulos. También queremos explorar las columnas que contienen diccionarios y por ultimo eliminar si tenemos valores duplicados en nuestros datos.

```python
import pandas as pd
import ast
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv("../data/working_data/idealista_data_cleaned_1.csv")
df
```

```python
df["num_columnas_nulas"] = df.isna().sum(axis=1)
df
```

```python
(df["num_columnas_nulas"] > 8).sum()
```

```python
df = df.drop(columns=["num_columnas_nulas"])
```

No hay demasiadas filas que tengan demasiados valores nulos

Todos los datos están bastantes completos 

```python
df.info()
```

```python
# Comenzamos con size, que vamos a poner los nulos a 0
df['size'] = df['size'].fillna(0)
```

```python
# Lo mismo lo hacemos con rooms y bathrooms
df['rooms'] = df['rooms'].fillna(0)
df['bathrooms'] = df['bathrooms'].fillna(0)
```

```python
# District y neighborhood los vamos a poner a "Unknown"
df['district'] = df['district'].fillna("Unknown")
df['neighborhood'] = df['neighborhood'].fillna("Unknown")
```

```python
df["parkingSpace"].head(20)
```

```python
# Convertimos los strings a diccionarios (solo si no son ya dicts)
df["parkingSpace"] = df["parkingSpace"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df["hasParkingSpace"] = df["parkingSpace"].apply(lambda x: x["hasParkingSpace"] if pd.notna(x) else None)
df["isParkingSpaceIncludedInPrice"] = df["parkingSpace"].apply(lambda x: x.get("isParkingSpaceIncludedInPrice") if pd.notna(x) else None)
```

```python
df.head(20)
```

Podemos ahora eliminar la columna de parkingSpace. Y lo que quiero comprobar ahora es cuantas filas tienen los valores de hasParkingSpace y isParkingSpaceIncludedInPrice a True

```python
conteo1 = ((df["hasParkingSpace"] == True) & (df["isParkingSpaceIncludedInPrice"] == True)).sum()

print(f"Número de filas donde ambas columnas son True: {conteo1}")
```

```python
conteo2 = ((df["hasParkingSpace"] == True) & (df["isParkingSpaceIncludedInPrice"] == False)).sum()

print(f"Número de filas donde hay parking pero no esta incluido en el precio: {conteo2}")
```

```python
conteo = ((df["hasParkingSpace"] == False)).sum()

print(f"Número de filas donde no hay parking: {conteo}")
```

```python
df["hasParkingSpace"].value_counts()
```

```python
df["isParkingSpaceIncludedInPrice"].value_counts()
```

```python
conteo3 = ((df["hasParkingSpace"] == True) & (df["isParkingSpaceIncludedInPrice"] != False) & (df["isParkingSpaceIncludedInPrice"] != True)).sum()
print(conteo3)
```

```python
conteo + conteo1 + conteo2 + conteo3
```

No se si es bueno o no quedarnos con las dos columnas en el modelo, la de saber si tiene parking o no disponible si que me parece buena. La de tener o no el precio incluido por ahora la voy a quitar

```python
df = df.drop(columns=["parkingSpace", "isParkingSpaceIncludedInPrice"])
```

```python
df["hasParkingSpace"] = df["hasParkingSpace"].fillna(False)
```

```python
# Los valores nulos en "priceByArea" los vamos a rellenar con la mediana de esa columna
mediana = df["priceByArea"].median()
df["priceByArea"] = df["priceByArea"].fillna(mediana)
```

```python
df["status"] = df["status"].fillna("unknown")
```
